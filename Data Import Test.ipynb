```python
# Import dependencies

import petpy as pt
from config import petfinder_api_key
from config import petfinder_secret
```

```python
# Initiate petpy

pf = pt.Petfinder(key=petfinder_api_key, secret=petfinder_secret)
```

```python
# Make API call

pet_data = pf.animals(animal_type='dog', status='adopted', results_per_page=100, pages=200, location=54701, before_date='2022-05-01', after_date='2021-05-01', return_df=True)
```
